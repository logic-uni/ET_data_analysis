## Load raw data

```python
import spikeinterface.full as si

import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
base_folder = Path(r'E:\xinchao\Data\useful_data\NP2\20250310_benchmark_control\Rawdata')

raw_rec = si.read_spikeglx(base_folder, stream_name='imec0.ap', load_sync_channel=False)
raw_rec
```

## Get LFP: Filter + Shift + CMR

```python
rec1 = si.bandpass_filter(recording=raw_rec, freq_min=0.2, freq_max=1000)
bad_channel_ids, channel_labels = si.detect_bad_channels(rec1)
rec2 = rec1.remove_channels(bad_channel_ids)
print('bad_channel_ids', bad_channel_ids)
rec3 = si.phase_shift(rec2)
rec = si.common_reference(rec3, operator="median", reference="global")
rec
```

# Save LFP

```python
job_kwargs = dict(n_jobs=40, chunk_duration='1s', progress_bar=True)

rec = rec.save(folder=base_folder / 'preprocess', format='binary', **job_kwargs)
```
